```python
import pandas as pd
```

```python
ds = pd.read_csv(r"C:\Users\ankit kumar jha\Downloads\data.csv")
print(ds.shape)
print(ds.columns)
print(ds.head())
```

```python
x=ds.drop('diagnosis',axis=1)
y=ds['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.36, random_state=42
)
```

```python
from sklearn.tree import DecisionTreeClassifier
c45 = DecisionTreeClassifier(
    criterion='entropy',        
    min_samples_split=10,       
    min_samples_leaf=5,         
    max_depth=6,                
    random_state=42
)

c45.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(35,20))
tree.plot_tree(c45, filled=True, feature_names=x.columns, class_names=["Benign", "Malignant"])
plt.show()
```

```python
y_pred = c45.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
print(score)
```
