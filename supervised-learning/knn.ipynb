```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
ds = pd.read_csv(r"C:\Users\ankit kumar jha\Downloads\archive (1)\drug200.csv")
print(ds.shape)

print(ds.head())
```

```python
x=ds.drop('Drug',axis=1)
y=ds['Drug']
```

```python
x=pd.get_dummies(ds,drop_first=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.36, random_state=42
)
```

```python
for k in range(1, 15):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"k={k}, accuracy={acc:.2f}")
```

```python
model = KNeighborsClassifier(n_neighbors=8)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}
```

```python
grid_knn = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5, scoring='accuracy')
grid_knn.fit(X_train, y_train)
```

```python
grid_knn.best_params_
```

```python

best_knn = grid_knn.best_estimator_

y_pred_best = best_knn.predict(X_test)
```

```python
print( accuracy_score(y_test, y_pred_best))
print( classification_report(y_test, y_pred_best))
```

```python
from sklearn.dummy import DummyClassifier

dummy = DummyClassifier(strategy="most_frequent")
dummy.fit(X_train, y_train)
y_dummy_pred = dummy.predict(X_test)
score=accuracy_score(y_dummy_pred,y_test)
print(score)
print(classification_report(y_test, y_dummy_pred))
```
